## Separando os 10 conjuntos de cada base

```
# SEPARANDO OS 10 CONJUNTOS DE CADA BASE #

folds = 10
kf = StratifiedKFold(n_splits=folds)


#Conjunto BASE GLASS/RECONHECIMENTO ÓTICO
x1_train = []
x1_test = []

y1_train = []
y1_test = []

#Conjunto BASE CARS
x2_train = []
x2_test = []

y2_train = []
y2_test = []


#Conjunto BASE ABALONE
x3_train = []
x3_test = []

y3_train = []
y3_test = []


for train_index, test_index in kf.split(X_1, y_1):
      x1_train.append(X_1[train_index])
      x1_test.append(X_1[test_index])

      y1_train.append(y_1[train_index])
      y1_test.append(y_1[test_index])


for train_index, test_index in kf.split(X_2, y_2):
      x2_train.append(X_2[train_index])
      x2_test.append(X_2[test_index])

      y2_train.append(y_2[train_index])
      y2_test.append(y_2[test_index])

  
for train_index, test_index in kf.split(X_3, y_3):
      x3_train.append(X_3[train_index])
      x3_test.append(X_3[test_index])

      y3_train.append(y_3[train_index])
      y3_test.append(y_3[test_index])


x1_train = list(x1_train)
x1_test = list(x1_test)

x2_train = list(x2_train)
x2_test = list(x2_test)

x3_train = list(x3_train)
x3_test = list(x3_test)

y1_train = list(y1_train)
y1_test = list(y1_test)

y2_train = list(y2_train)
y2_test = list(y2_test)

y3_train = list(y3_train)
y3_test = list(y3_test)


# print('\n')
# for i in range(10):
#   print(len(y3_test[i]))
```

## Árvore de Decisão

```
from sklearn.model_selection import cross_val_score
from sklearn import tree


def DecisionTree(x_train, x_test, y_train, y_test, criterion=""):
      model = tree.DecisionTreeClassifier(criterion=criterion)
      model = model.fit(x_train, y_train)
      results =[]

      result = model.predict(x_test)
      acc = metrics.accuracy_score(result, y_test)
      results.append(acc)

      # print(results)
      show = round(np.mean(results) * 100)
      # print("{}%".format(show))  

      final_results = {
          "results": results,
          "medium_percentage_acc": show,
      }

      return final_results

```

# KNN

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

def KNN(x_train, x_test, y_train, y_test, k):
     model = KNeighborsClassifier(n_neighbors=k, metric='euclidean', algorithm='brute')

     model = model.fit(x_train, y_train)
     results = []
     
     result = model.predict(x_test)
     acc = metrics.accuracy_score(result, y_test)
     results.append(acc)

     show = round(np.mean(results) * 100)

     final_results = {
          "results": result,
          "medium_percentage_acc": show,
      }

     return final_results
```

# MLP

```
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

def MLP(x_train, x_test, y_train, y_test, architeture=1,  activation='', plot=False):
    if (architeture == 1):
      model = MLPClassifier(hidden_layer_sizes=(9, 6, 3), activation=activation, max_iter=3000, random_state=1)
    else:
      model = MLPClassifier(hidden_layer_sizes=(7, 5, 3), activation=activation, max_iter=3000)

    model = model.fit(x_train, y_train)
    results = []

    if plot == True:
      plt.plot(model.loss_curve_)
      plt.show()


    result = model.predict(x_test)
    acc = metrics.accuracy_score(result, y_test)
    results.append(acc)

    show = round(np.mean(results) * 100)

    final_results = {
          "results": result,
          "medium_percentage_acc": show,
      }

    return final_results
```

# K-MEANS

```
from sklearn.cluster import KMeans
from collections import Counter

def K_Means(x_train, x_test, y_train, y_test):
  myset = set(y_train) # Cria um conjunto. Em conjuntos, dados não se repetem. Assim, esse conjunto conterá apenas um valor de cada, ou seja: [1,2,3]
  clusters = len(myset) # Quantos clusters teremos no KMeans

  model = KMeans(n_clusters = clusters)
  model = model.fit(x_train)

  labels = model.labels_

  # ### Plot do KMeans
  # y_kmeans = model.predict(x_train)
  # X, y_true = make_blobs(n_samples=300, centers=clusters, cluster_std=0.60, random_state=0)
  # plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
  # centers = model.clusters_centers_ 
  # plt.scatter(centers[:, 0], centers[:, 1], s=200, alpha=0.5);

  map_labels = []

  for i in range(clusters):
    map_labels.append([])
  
  new_y_train = list(y_train)

  for i in range (len(y_train)):
    for c in range(clusters):
      if labels[i] == c:
        map_labels[c].append(new_y_train[i])

  ## Criar dicionário com os labels a serem mapeados
  mapping = {}

  for i in range(clusters):
      final = Counter(map_labels[i])  # contar a classe que mais aparece
      value = final.most_common(1)[0][0]  # retorna a classe com maior frequência
      mapping[i] = value

  result = model.predict(x_test)
  result = [mapping[i] for i in result]

  acc = metrics.accuracy_score(result, y_test)
  show = round(acc * 100)

  final_results = {
          "results": result,
          "medium_percentage_acc": show,
      }

  return final_results
```
